```python
import pandas as pd
import os

#due to the whole processing being time intensive it was done beforehand and saved in
#train_processed3.csv file
if processed := os.path.isfile("data/train_processed.csv"):
    data = pd.read_csv("data/train_processed.csv")
else:
    data = pd.read_csv("data/train.csv")
    
rows = data.shape[0]
cols = data.shape[1]
collist = data.columns
collist
```

```python
import numpy as np
import os

if not processed:
    for c in collist[10:-1]:
        for r in range(rows):
            if not data.loc[r,c] in [np.nan, 'nan', 0.0]:
                try:
                    data.loc[r,c] = float(data.loc[r,c][6:-6])
                except:
                    print("exception at",r,c)
    #now to fill null values with mean
#     data.fillna(data.mean())
data
```

```python
# data = data.drop(list(collist[:10]),axis=1)
y = data["state"]
x = data.drop("state",axis=1)

x = x.fillna(x.mean())

print(x.columns)
#check:
# for col in x.columns:
#     print(col,": ",x[col].isna().sum())
```

```python
from sklearn.preprocessing import LabelEncoder as LE
from sklearn.model_selection import train_test_split

le = LE()
y = le.fit_transform(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
print(len(x_train),len(y_train),len(x_test),len(y_test))
```

```python
from sklearn.linear_model import LogisticRegression as LogReg
lr = LogReg(max_iter=4000,C=100)
lr.fit(x_train,y_train)
```
