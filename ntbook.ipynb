```python
import pandas as pd
import os

#due to the whole processing being time intensive it was done beforehand and saved in
#train_processed3.csv file
if processed := os.path.isfile("data/train_processed.csv"):
    data = pd.read_csv("data/train_processed.csv")
else:
    data = pd.read_csv("data/train.csv")
    
rows = data.shape[0]
cols = data.shape[1]
collist = data.columns
collist
```

```python
import numpy as np
import os

def proc(data,test=False):
    collist = data.columns[10:] if test else data.columns[10:-1]
    for c in collist:
        for r in range(data.shape[0]):
            if not data.loc[r,c] in [np.nan,'nan',0.0]:
                try:
                    data.loc[r,c] = float(data.loc[r,c][6:-6])
                except:
                    print("exception at",r,c)
        print(f"col : {col} done")
    return data

if not processed:
    data = proc(data)
#     for c in collist[10:-1]:
#         for r in range(rows):
#             if not data.loc[r,c] in [np.nan, 'nan', 0.0]:
#                 try:
#                     data.loc[r,c] = float(data.loc[r,c][6:-6])
#                 except:
#                     print("exception at",r,c)
    #now to fill null values with mean
#     data.fillna(data.mean())
data
```

```python
data = data.drop(list(collist[:10]),axis=1)
y = data["state"]
x = data.drop("state",axis=1)

x = x.fillna(x.mean())

print(x.columns)
#check:
print(y)
for col in x.columns:
    print(col,": ",x[col].isna().sum())
```

```python
from sklearn.preprocessing import OneHotEncoder as OH
import tensorflow as tf
from sklearn.model_selection import train_test_split

y = y.values.reshape(-1,1)
enc = OH(sparse_output=False)
y = enc.fit_transform(y)
# le = LE()
# y = le.fit_transform(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
print(len(x_train),len(y_train),len(x_test),len(y_test))
print(x_train.shape,x_train)
x_train[collist[10]]
```

```python
# x_train = (x_train-x_train.min())/(x_train.max()-x_train.min())
# x_train
```

```python
inp_shape = x_train.shape[1]
out_shape = y_train.shape[1]

model = tf.keras.Sequential([
    tf.keras.Input(shape=(inp_shape,)),
    tf.keras.layers.Dense(80,activation='relu'),
    tf.keras.layers.Dense(75,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(15,activation='relu'),
#     tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(out_shape, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
batch_size = 8
epochs = 10
model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test loss: {loss}, Test accuracy: {accuracy}')
```
